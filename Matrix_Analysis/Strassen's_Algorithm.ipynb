### Q2. Standard matrix multiplication

```python
# Standard case: assumes n=2^k and matrices are square
# This code is written for transparency and is 
# highly sub-optimal both for memory usage and speed.
def stmm(A: np.ndarray, B: np.ndarray) -> np.ndarray:

    n=A.shape[0]

    if n==2: 
        return A @ B                     

    hn=n // 2 # half size of current block

    A11=A[:hn, :hn]
    A12=A[:hn, hn:]
    A21=A[hn:, :hn]
    A22=A[hn:, hn:]

    B11=B[:hn, :hn]
    B12=B[:hn, hn:]
    B21=B[hn:, :hn]
    B22=B[hn:, hn:]

    # recursively call the same function
    C11=stmm(A11, B11) + stmm(A12, B21)
    C12=stmm(A11, B12) + stmm(A12, B22)
    C21=stmm(A21, B11) + stmm(A22, B21)
    C22=stmm(A21, B12) + stmm(A22, B22)

    # reassemble
    C=np.block([[C11, C12],
                  [C21, C22]])

    return C

n=16
A=np.random.randn(n,n)*1.0
B=np.random.randn(n,n)*1.0
C_stmm=stmm(A,B) # using function               
C_np=A@B # from numpy                    

# Verify 
print("Maximum absolute difference:", np.max(np.abs(C_stmm - C_np)))
```

### Q3. Strassen's algorithm for matrix multiplication

```python
# Strassen algorithm: assumes n=2^k and matrices are square.
# This code is written for transparency and is 
# highly sub-optimal both for memory usage and speed.
def samm(A: np.ndarray, B: np.ndarray) -> np.ndarray:

    n=A.shape[0]

    if n==2: # ordinary multiplication at n=2 level
        return A @ B                     

    hn=n // 2                         # half‑size of the current block

    A11=A[:hn, :hn]
    A12=A[:hn, hn:]
    A21=A[hn:, :hn]
    A22=A[hn:, hn:]

    B11=B[:hn, :hn]
    B12=B[:hn, hn:]
    B21=B[hn:, :hn]
    B22=B[hn:, hn:]

    # recursively call the same function
    M1=samm(A11 + A22, B11 + B22)
    M2=samm(A21 + A22, B11)
    M3=samm(A11, B12 - B22)
    M4=samm(A22, B21 - B11)
    M5=samm(A11 + A12, B22)
    M6=samm(A21 - A11, B11 + B12)
    M7=samm(A12 - A22, B21 + B22)

    # combine
    C11=M1 + M4 - M5 + M7
    C12=M3 + M5
    C21=M2 + M4
    C22=M1 - M2 + M3 + M6

    # reassemble
    C=np.block([[C11, C12],
                  [C21, C22]])

    return C

n = 16
A = np.random.randn(n, n) * 1.0
B = np.random.randn(n, n) * 1.0

C_samm = samm(A, B)                # result from our Strassen routine
C_ref  = A @ B                     # reference result from NumPy

# Verify 
diff = np.max(np.abs(C_samm - C_ref))
print(f"Maximum absolute difference vs. NumPy matmul: {diff:.2e}")
```
