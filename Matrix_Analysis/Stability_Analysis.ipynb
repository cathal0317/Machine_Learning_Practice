### Q3d. Stable versus unstable versions

```python
# set up the two versions of the function
def f(x): # unstable version
    return np.sqrt(1 + x**2) - 1
def g(x): # stable version
    # algebraic reformulation that avoids catastrophic cancellation
    return x**2 / (np.sqrt(1 + x**2) + 1)

# get a range of x values
x=10.0**np.arange(-8,-4+0.2,0.2)  

# plot data
plt.figure(figsize=(10, 4))
plt.semilogx(x, f(x) / g(x), ':o')
plt.ylabel('Unstable / Stable solution')
plt.xlabel('x');
```

```python
# Machine precision is around 10^{-16} for float64
# and the unstable code breaks down between
# 10^{-8} and 10^{-7} which fits in with 
# roughly epsilon_m/x^2 being the divergent term when x is small.
```
