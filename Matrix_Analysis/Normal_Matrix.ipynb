### Q2a. p-norm of a complex vector

```python
import numpy as np

def my_p_norm(z, p):
    az=np.abs(z)
    if p==np.inf:
        return np.max(az)
    else:
        return np.sum(az**p)**(1.0/p)

# Check
z = np.array([1+2j, 2, 5j])
print(my_p_norm(z,1),my_p_norm(z,2),my_p_norm(z,np.inf))
```

### Q2b. Check for unitarity

```python
def my_is_unitary(M, rtol=1e-12, atol=1e-14):
    
    X = M.conj().T @ M           
    I = np.eye(M.shape[0], dtype=M.dtype)
    ok = np.allclose(X, I, rtol=rtol, atol=atol)
    return X,ok

# Unitary Example
ang = 2*np.pi/3
U = np.array([[1, 1, 1],
              [1, np.exp(-1j*ang), np.exp(-2j*ang)],
              [1, np.exp(-2j*ang), np.exp(-4j*ang)]],
              dtype=complex) / np.sqrt(3)

# Non unitary
#U=np.random.rand(3,3)

X_u, ok_u = my_is_unitary(U)
print(ok_u)          # True
```

### Q2c. Check for normality

```python
def my_is_normal(A, rtol=1e-12, atol=1e-14):

    A_H=A.conj().T
    X=A_H @ A
    Y=A @ A_H
    is_normal=np.allclose(X, Y, rtol=rtol, atol=atol)
    return X, Y, is_normal

# Example
A=np.array([
    [2, 1+1j, 0],
    [1-1j, 3, 0],
    [0, 0, 4]
], dtype=complex)

A=np.random.rand(3, 3)

X_norm, Y_norm, c_norm = my_is_normal(A)
print(f"Normal Matrix Check: Is it normal? {c_norm}")
```
