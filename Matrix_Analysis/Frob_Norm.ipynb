```python
import numpy as np
```

```python
# Q2a
A=np.random.randn(3, 2)

# Q2b and Q2c
Q,R=np.linalg.qr(A,mode='reduced')  

# Q2d 
def my_frobenius_norm(M: np.ndarray) -> float:
    m,n=M.shape
    acc=0.0
    for i in range(m):
        for j in range(n):
            acc += M[i, j] ** 2
    return np.sqrt(acc)

residual=A - Q@R             
E_custom=my_frobenius_norm(residual)

print(f"E = {np.format_float_scientific(E_custom, precision=2)}")
```
